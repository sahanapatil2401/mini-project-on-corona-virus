# positive cases according to state

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(11, 8), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states,df['positive'], align='center', alpha=0.5,color=(1,0,0),edgecolor=(0.5,0.2,0.8))
plt.yticks(total_states, df['state_name'])
plt.xlim(1,max(df['positive'])+100)
plt.xlabel('Positive Number of Cases')
plt.title('Corona Virus Cases')
plt.show()
```

active case according to state

```python
from matplotlib.pyplot import figure
figure(num=None,figsize=(11,9),dpi=80,facecolor='w',edgecolor='r')
plt.barh(total_states,df['new_active'],align='center',alpha=0.9,color=(1,0,0),edgecolor=(0.2,0.9,0.7))
plt.yticks(total_states,df['state_name'])
plt.xlim(1,max(df['new_active'])+100)
plt.xlabel("active cases")
plt.title('corona active case')
plt.show()
```

death case according to states

```python
from matplotlib.pyplot import figure
figure(figsize=(9, 8))
plt.barh(total_states,df['death'],align='center')
plt.yticks(total_states,df['state_name'])
plt.xlim(1,max(df['death'])+100)
plt.xlabel("death cases")
plt.title ("corona virus death case")
plt.show()
```
